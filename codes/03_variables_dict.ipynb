````python
# dictionary 안에 key name은 ""으로 감싸준다.
# dictionary는 {}로 감싸서 배열을 넣어준다.

first = { "second" : "good" }
````

````python
first
````

````python
first["second"]
````

````python
second = {
    "first" : "good",
    "second" : "bad"
}
````

````python
second["second"]
````

````python
third = "not bad"
````

````python
second = {
    "first" : "good",
    "second" : "bad",
    "not_key" : third
}
````

````python
second["not_key"]
````

### 문제 2. 딕셔너리 값 접근하기  
다음과 같은 딕셔너리가 있습니다.  
`"city"`에 해당하는 값을 출력하는 코드를 작성해 보세요.

```python
info = {
    "name": "Bob",
    "age": "25",
    "city": "Seoul"
}
# "city" 값을 출력
```

````python
info = {
    "name": "Bob",
    "age": "25",
    "city": "Seoul"
}
print(info["city"])
````
